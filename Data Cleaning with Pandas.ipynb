```python
import pandas as pd
```

```python
df=pd.read_excel('/content/Patient Demographic information.xlsx')
```

```python
df.head()
```

```python
#Drop any duplicate records

df = df.drop_duplicates()
df
```

```python
#drop the last column as it does not add any value

df=df.drop(columns = "Default column")
df
```

```python

#remove numeric and special characters from the frist and last name
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
#clean phone number column.Remove special characters
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
# add - as a proper phone number format
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python

#split address in to  street,city and zip code level
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',2, expand=True)
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df=df.fillna('')
df
```

```python

#remove all the rows where do not contact =y
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df
```

```python
df = df.reset_index(drop=True)
df
```
