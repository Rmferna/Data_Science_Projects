Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis.PIMA Dataset-Keggle.com

```python
diabetes_dataset=pd.read_csv('/content/diabetes.csv')
```

```python
#first 5 rows
diabetes_dataset.head()
```

```python
# number of columns and rows
diabetes_dataset.shape
```

```python
#statistial measures
diabetes_dataset.describe()
```

```python

diabetes_dataset['Outcome'].value_counts()
```

0 = No Diabetic

1 = Diabetic




```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
#seperate data an label
X=diabetes_dataset.drop(columns='Outcome',axis=1)
Y=diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Data Standerization-Even though the metadata is all numeric,they are in different standards.eg-Glucose is in 100s BMI is in float.

```python
#use standard scaler function which we import in the begining.
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data=scaler.transform(X)
```

```python
print(standardized_data)
```

```python
#X will represent the data and Y will represent the mode
X=standardized_data
Y=diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

Train Test Split

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Training the model.

```python
classifier=svm.SVC(kernel='linear')
```

```python
#training support vector machin classifier
classifier.fit(X_train,Y_train)
```

Model Evaluation

```python
#Accuracy Score on training data
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy score of the training data : ',training_data_accuracy)
```

```python
#Accuracy Score on test data
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score of the test data : ',test_data_accuracy)
```

Training data accuracy and Test data accuracy alls with in same rage,which means no overfitting.

Built Predictive System

```python
input_data=(10,168,74,0,0,38,0.537,34)

#convert the input data in to a numpy array
input_data_as_numpy_array=np.asarray(input_data)

#reshape the array as we are predicting for one instance
#(reason for reshape is to inform the model that we are going to predic only one instance)
input_data_reshape=input_data_as_numpy_array.reshape(1,-1)

#standerdize data
std_data=scaler.transform(input_data_reshape)

print(std_data)
prediction=classifier.predict(std_data)
print(prediction)
if (prediction[0]==0):
  print('Negative')
else:
  print('Positive')
```
